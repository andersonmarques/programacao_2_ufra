```python
import plotly.graph_objects as go
import squarify

# Dados
labels = ['Matriculados', 'Aprovados', 'Reprovados']
quantidades = [13, 12, 1]
porcentagens = [q / quantidades[0] * 100 for q in quantidades]

# Cores das categorias
cores = ['#ff9999', '#66b3ff', '#99ff99']

# Calcula os tamanhos dos retângulos no formato squarified
sizes = squarify.normalize_sizes(quantidades, 100, 100)

# Criação do treemap
fig = go.Figure(go.Treemap(
    labels=labels,
    parents=['', 'Matriculados', 'Matriculados'],
    values=sizes,
    text=[f'{labels[i]} - {porcentagens[i]:.2f}%' for i in range(len(labels))],  # Concatena o nome da categoria com a porcentagem
    hovertemplate='<b>%{label}</b>',
    texttemplate='%{text}',
    marker=dict(colors=cores),
    textfont=dict(size=15),
))

# Configurações adicionais do treemap
fig.update_layout(
    title='Distribuição dos Discentes Matriculados',
    margin=dict(t=100, b=0),
)

# Exibe o treemap
fig.show()
```
